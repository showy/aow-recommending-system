```python
from surprise.prediction_algorithms.matrix_factorization import SVDpp
from surprise import Dataset
from surprise import evaluate, print_perf
from surprise import GridSearch
import pandas as pd
from surprise import Reader
from sklearn.metrics import precision_score, recall_score, roc_auc_score
from multiprocessing import Pool
from sklearn.externals import joblib
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_file = "../datasets/movielens/ml-100k/ua.base"
test_file = "../datasets/movielens/ml-100k/ua.test"
```

```python
col_names = [ "user", "movie", "rating", "timestamp" ]
```

```python
train = pd.read_csv(train_file, sep="\t", header=None, names=col_names)
train['rating'] = 1
```

```python
test = pd.read_csv(test_file, sep="\t", header=None, names=col_names)
test['rating'] = 1
```

# Build model


```python
reader = Reader(rating_scale=(0, 1))
```

```python
train_dataset = Dataset.load_from_df(train[["user", "movie", "rating"]], reader).build_full_trainset()
```

```python
svdpp = SVDpp()
```

```python
svdpp.fit(train_dataset)
```

```python
test_dataset = Dataset.load_from_df(test[["user", "movie", "rating"]], reader).build_full_trainset()
```

```python
predictions = svdpp.test(test_dataset.build_testset())
```
